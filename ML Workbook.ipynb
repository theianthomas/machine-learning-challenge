```python
import pandas as pd
```

```python
planets_raw = pd.read_csv("cumulative.csv")
planets_1st = planets_raw.copy().drop(columns=["rowid", "kepid", "kepoi_name", "kepler_name", "koi_pdisposition", "koi_score", "koi_tce_delivname"])
# Drop the null columns where all values are null
planets_2nd = planets_1st.dropna(axis='columns', how='all').dropna().reset_index(drop=True).copy()
# Drop the null rows
planets_2nd.sample(10)
```

```python
# Count number of times each value in koi_disposition occurs
# There is a disparity between FALSE POSITIVE and the other targets.
# This effects model performace in favor of ML arriving at a model that determines FALSE POSITIVE more frequently
planets_2nd["koi_disposition"].value_counts()
```

```python
# Filter columns to look for pattern in specific columns
planets_2nd.filter(regex='(disposition|_nt|ss|co|_ec)').groupby("koi_disposition").sum()
```

```python
# Apply filter to examine columns having to do with _err
planets_2nd.filter(regex='(disposition|err)').groupby("koi_disposition").mean()
```

```python
# Apply filter to examine columns not containing _err

planets_2nd.filter(regex = '^((?!err).)*$').groupby("koi_disposition").mean()
```

```python
# Filtered out err columns to reduce data size for processing.
planets_filter_err = planets_2nd.filter(regex = '^((?!err).)*$')
planets_filter_err.head()
```

```python
#voice = pd.read_csv('../Resources/voice.csv')

X = planets_2nd.drop("koi_disposition", axis=1)
y = planets_2nd["koi_disposition"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.utils import to_categorical
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Create a StandardScaler model and fit it to the training data
X_scaler = MinMaxScaler().fit(X)

# Transform the training and testing data using the X_scaler

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

print(X_train_scaled)
print(X_test_scaled)
```

```python
y_test.head()
```

```python
from sklearn import preprocessing
#from keras.utils import to_categorical

le = preprocessing.LabelEncoder()
le.fit(y)

#LabelEncoder()
#Encode classes for target data

y_train_encoded = le.transform(y_train)
y_test_encoded = le.transform(y_test)

print(le.classes_)
print(y_test_encoded)
print(y_train_encoded)
```

```python
from keras.utils import to_categorical


# One-hot encoding
y_train_categorical = to_categorical(y_train_encoded)
y_test_categorical = to_categorical(y_test_encoded)

print(y_test_categorical)
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC 
from sklearn.metrics import classification_report
```

```python
# Calculate classification report using lables.
model_SVC_scaled = SVC(kernel='linear')
model_SVC_scaled.fit(X_train_scaled, y_train)
prediction_SVC_scaled = model_SVC_scaled.predict(X_test_scaled)


print(classification_report(y_test, prediction_SVC_scaled,
                            target_names=['CANDIDATE', 'CONFIRMED', 'FALSE POSITIVE']))
```

```python
# Calculate classification report using encoded values to determine if numerical values change SVC outcome.
model_SVC_encoded = SVC(kernel='linear')
model_SVC_encoded.fit(X_train_scaled, y_train_encoded)
prediction_SVC_encoded = model_SVC_encoded.predict(X_test_scaled)


print(classification_report(y_test_encoded, prediction_SVC_encoded,
                            target_names=['CANDIDATE', 'CONFIRMED', 'FALSE POSITIVE']))
```

```python
print(f"Training Data Score: {model_SVC_encoded.score(X_train_scaled, y_train_encoded)}")
print(f"Testing Data Score: {model_SVC_encoded.score(X_test_scaled, y_test_encoded)}")
```

```python
print(f"Training Data Score: {model_SVC_scaled.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model_SVC_scaled.score(X_test_scaled, y_test)}")
```

```python
#### Conclusion
#The f1-score seems the model predicts False Positive more reliably. This could be due to the disproportionate representation of False Positive data.
#Encoding the target values has created no difference in the output score. SVC models 
```

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10, 50],
              'gamma': [0.0001, 0.0005, 0.001, 0.005]}
grid_SVC = GridSearchCV(model_SVC_scaled, param_grid, verbose=3)
```

```python
# List the best parameters for this dataset
print(grid_SVC.best_params_)
```

```python
# List the best score
print(grid_SVC.best_score_)
```

```python
# first, create a normal neural network with 2 inputs, 6 hidden nodes, and 2 outputs
from keras.models import Sequential
from keras.layers import Dense

model_node = Sequential()
model_node.add(Dense(units=120, activation='relu', input_dim=40))
model_node.add(Dense(units=3, activation='softmax'))
```

```python
model_node.summary()
```

```python
# Compile the model
model_node.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit the model to the training data
model_node.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=10,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model_node.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
#### Conclusion
#With very little training the neural achieves the same accuracy as the SVM model. However with the Hypertuning the SVM model achieves significantly higher accuracy of 88% which surpasses the Neural Network. It would be interesting to see if/how-many epochs of the neural network it would take to achieve the same accuracy. 
```
